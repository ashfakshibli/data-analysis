#### Which country did not win a medal in 2008 and how many countries.

```python
ol2008 = oo[oo.Edition == 2008]
ol2008.head()
```

```python
noc = pd.read_csv('data/Summer Olympic medallists 1896 to 2008 - IOC COUNTRY CODES.csv')
noc.head()
```

```python
noc[noc['Country'] != noc['Country.1']]
```

```python
noc.set_index('Int Olympic Committee code', inplace=True)
noc.head()
```

```python
medal_2008 = ol2008.NOC.value_counts()
medal_2008
```

```python
noc['medal_2008'] = medal_2008
noc.head()
```

```python
noc[noc.medal_2008.isnull()].count()
```

```python
noc[noc.medal_2008.isnull()]
```
