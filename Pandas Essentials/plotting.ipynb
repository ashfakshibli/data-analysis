# Basic Plotting

```python
import pandas as pd
oo = pd.read_csv("data/olympics.csv", skiprows=4)

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
oo.head()
```

## Plot first edition of Olympics.

```python
fo = oo[oo.Edition == 1896]
fo.head()
```

```python
fo.Sport.value_counts().plot(kind='line')
```

```python
fo.Sport.value_counts().plot(kind='bar')
```

```python
fo.Sport.value_counts().plot(kind='barh')
```

```python
fo.Sport.value_counts().plot(kind='pie')
```

```python
fo.Sport.value_counts().plot(kind='line', color='maroon', figsize=(10,3))
```

### Color Maps
#### Sequential, Diverging, Qualitative

```python
fo.Sport.value_counts().plot(kind='pie', colormap='Pastel1')
```

### Seaborn Plotting

```python
import seaborn as sns
```

#### How many medals have been won by men and women in the history of Olympics? Gold, Silver, Bronze wise data also needed.

```python
sns.countplot(x='Medal', data=oo, hue='Gender') # hue for categorical variable
```

### Plot number of medals achieved by USA (men and women) in Beijing 2008.

```python
us = oo[(oo.NOC == 'USA') & (oo.Edition == 2008)]
us.head()
sns.countplot(x='Medal', data=us,hue='Gender', palette='bwr', order=['Bronze', 'Silver', 'Gold'])
```
